```python
import pandas as pd
```

```python
df = pd.read_csv("Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv")

df = df[~df["LocationDesc"].isin(["Virgin Islands", "Puerto Rico", "Guam"])]
```

```python
df.to_csv('nutri_data.csv')
```

```python
df['Question'].unique()
```

```python

df = df.pivot_table(
    index=["LocationDesc", "Question"],  # Rows: Unique combination of Location and Question
    columns=["StratificationCategory1"],       # Columns: Unique Stratification categories
    values="Data_Value",              # Values to fill in the table
    aggfunc="mean"                  # Aggregation function if there are duplicates
).reset_index()
```

```python
df1 = df[df['Question'] == 'Percent of adults aged 18 years and older who have obesity'][['LocationDesc', 'Total']]
df2= df[df['Question'] == 'Percent of adults who engage in muscle-strengthening activities on 2 or more days a week'][['LocationDesc', 'Total']]
df3 = 
combined = pd.merge(df1,df2, on= 'LocationDesc')

combined['Total_x'].corr(combined['Total_y'])
```

```python
df_total = df[df['Stratification1'] == 'Total']

# Pivot so the Question is a column, the average over each year
pivoted_df = df_total.pivot_table(index="LocationDesc", columns="Question", values="Data_Value", aggfunc="mean")
```

```python
pivoted_df.corr()
```

```python
# Reshaping the DataFrame so the data is in organized columns
reshaped_df = df.pivot_table(
    index=["LocationDesc", "Question", "YearStart"],  # Rows: Unique combination of Location and Question
    columns="StratificationCategory1",       # Columns: Unique Stratification categories
    values="Data_Value",              # Values to fill in the table
    aggfunc="mean"                  # Aggregation function if there are duplicates
).reset_index()
```

```python
# Filter the DataFrame to exclude Virgin Islands, Puerto Rico, and Guam
filtered_df = reshaped_df[~reshaped_df["LocationDesc"].isin(["Virgin Islands", "Puerto Rico", "Guam"])]
```
